Generate the YAML file to direct the YOLOv7 trining file.

```python
classes_fl = open("/content/yolo_data/classes.txt", "r")
classes = classes_fl.read()
classes_fl.close()
classes = classes.split("\n")[:-1]

class_str = '['
for cls in classes:
  class_str += ("'" + cls + "', ")

class_str = class_str[:-2] + "]"
nc = len(classes)
```
